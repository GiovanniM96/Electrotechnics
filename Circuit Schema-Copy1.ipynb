# Circuit Schema

```python
import schemdraw
import schemdraw.elements as elm

d = schemdraw.Drawing()

V1 = d.add(elm.SOURCE_V, d='up', label='$V_{in}$')
R1 = d.add(elm.RES, d='right', label='R')
C1 = d.add(elm.CAP, d='down', botlabel='C')
GND = d.add(elm.GND)

d.add(elm.LINE, start=V1.start, end=R1.start)
d.add(elm.LINE, start=R1.end, end=C1.start)
d.add(elm.LINE, start=C1.end, end=GND.start)
    

d.draw()
```

```python
with schemdraw.Drawing() as d:
    d.add(elm.Resistor())
    d.add(elm.Capacitor())
    d.add(elm.Diode())
```

```python
with schemdraw.Drawing() as d:
    d += elm.Resistor()
    d += elm.Capacitor()
    d += elm.Diode()
```

```python
with schemdraw.Drawing() as d:
    d += elm.Resistor().label('100KÎ©')
    d += elm.Capacitor().down().label('0.1Î¼F', loc='bottom')
    d += elm.Line().left()
    d += elm.Ground()
    d += elm.SourceV().up().label('10V')
```
